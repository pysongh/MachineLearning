```python
'''
JJOLAB ML01 : Linear regression 20171120
Linear regression을 통해 아래의 내용을 이해한다.
    1. Error 함수를 이해
    2. Gradient Descent 를 이해
    3. Linear regression을 이해
'''

#초기값 설정
init_w = 0 # weight 초기값
init_b = 0 # bias 초기값
learning_rate = 0.003 # 학습속도 test 0.003 -> 0.0001 -> 0.00001
N = len(x) # 데이터 수
step = 1000 # 반복학습 횟수
```

```python
# error를 정의
def compute_error(current_w, current_b, x, y):
    totalError = 0
    # data 수 반큼 반복한다
    for i in range(N):
        totalError += 1/N * (y[i] - (current_w * x[i] + current_b))**2
    return totalError
```

```python
# gradient descent를 정의
def gradient_descent(current_w, current_b, x, y, learningRate):
    #new_w = current_w - r*gradient(w)
    #new_b = current_w - r*gradient(b)
    for i in range(N):
        gradient_w = - 2/N * x[i] * (y[i] - (current_w * x[i]+ current_b)) 
        gradient_b = - 2/N * (y[i] - (current_w * x[i]+ current_b))
    new_w = current_w - learningRate * gradient_w
    new_b = current_b - learningRate * gradient_b    
    return new_w, new_b
```

```python
w, b = init_w, init_b

for i in range(step):
    # calculate current error
    error = compute_error(w,b,x,y)
    # print current status
    print("step : {}, w : {:.3f}, b : {:.3f}, error: {:.1f}".format(i,w,b,error))
    # update w,b
    w, b = gradient_descent(w,b,x,y,learning_rate)
```
